## Ejercicio 3 Filtrar atípicos y manejar datos ausentes:
•	Realiza un análisis estadístico o inspección visual de cada columna numérica para identificar valores atípicos.
•	Usa Pandas para filtrar y eliminar los datos atípicos y los valores ausentes. Explica en tu entrega qué criterios utilizaste para identificar los atípicos.



```python
# 1. Seleccionar solo columnas numéricas, generó una lista con el nombre de las columnas con valores numericos.
columnas_numericas = df_total.select_dtypes(include='number').columns

# 2. Creamos una máscara booleana que marca las filas que NO tienen valores atipicos
filtro_valores_no_atipicos = pd.Series([True] * len(df_total))

# 3. Recorremos cada columna numérica 
for i in columnas_numericas:
    Q1 = df_total[i].quantile(0.25)
    Q3 = df_total[i].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    # 4. Creamos una máscara para esta columna (True si está dentro de los límites)
    mascara_columna = (df_total[i] >= limite_inferior) & (df_total[i] <= limite_superior)
    
    # 5. Actualizamos la máscara general (conserva solo las filas sin valores atipicos en ninguna columna)
    filtro_valores_no_atipicos &= mascara_columna

# 6. Filtramos el DataFrame lo hacemos con copy para no crear dependencias
df_total_limpio = df_total[filtro_valores_no_atipicos].copy()

# 7. Resultado
print(f"Filas originales: {len(df_total)}")
print(f"Filas después de eliminar valores atipicos: {len(df_total_limpio)}")
```
